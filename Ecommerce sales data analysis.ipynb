# Customer satisfaction prediction on ecommerece data of Olist

## 1. Business problem, and mapping it to Data/Machine learning problem

## 1.1 Description

Olist is an e-commerce site of Brazil which provides a better platform to connect merchants and their product to the main marketplace of Brazil. Olist released this dataset on kaggle in Nov 2018. The data-set has information of 100k orders from 2016 to 2018 made at multiple marketplaces in Brazil. Its features allow viewing an order from multiple dimensions: from order status, price, payment and freight performance to customer location, product attributes and finally reviews written by customers. A Geo-location data-set that relates Brazilian zip codes to lat/long coordinates has also been released.

Olist provides good services for the merchants to sell their products through the Olist Store and ship them directly to the customers using Olist logistics partners.Its business is based on interaction between consumer ,olist store and the seller.At first an order is made by the consumer on the olist site.This order is received by olist store ,based on the information of the order (like product category,geolocation,mode of payment e.t.c) a notification is forwarded to the sellers. After that product is received from the seller and delivered to the consumer within the estimated delivery time. Once the customer receives the product, or if the estimated delivery date is due, the customer gets a satisfaction survey by email where he can give a note for the purchase experience and write down some comments.

## 1.2 Problem statement

- For given historical data, predict review score for next order.
- This problem can further modified to, predicting customer satisfaction as positive (1) or negative (0) for the purchase they made on Olist

```python
temp = pd.DataFrame(df.groupby('payment_type')['review_score'].agg(lambda x: x.eq(1).sum())).reset_index()

temp['total'] = list(df.groupby('payment_type')['review_score'].agg('count'))

temp['avg']   = list(df.groupby('payment_type')['review_score'].agg('mean'))

temp = temp.sort_values(by=['total'], ascending=True)
temp.reset_index(drop=True)
```

```python
def pareto_plot(df, x, y, title=None, show_pct_y=False, pct_format='{0:.0%}'):
    
    tmp = df.sort_values(y, ascending=False)
    
    x_values = tmp[x].values
    y_values = tmp[y].values
    
    weights = y_values/sum(y_values)
    cumsum = weights.cumsum()
    
    fig, ax1 = plt.subplots(figsize=(10,6))
    
    ax1.bar(x_values, y_values)
    ax1.set_xlabel(x)
    ax1.set_ylabel(y)
    
    ax2 = ax1.twinx() # creates secondary y-axis on base plot obj, and returns a new obj 
    ax2.plot(x_values, cumsum, "-ro")
    ax2.set_ylabel('', color='r')
    ax2.tick_params('y', colors='r')
    
    vals = ax2.get_yticks()
    print(vals)
    ax2.set_yticklabels(['{:,.2%}'.format(x) for x in vals])
    
    if not show_pct_y:
        ax2.set_yticks([])
    
    formatted_weights = [pct_format.format(x) for x in cumsum]
    for i, txt in enumerate(formatted_weights):
        ax2.annotate(txt, (x_values[i], cumsum[i]),fontsize=15)
        
    plt.tight_layout()
    plt.show()
```

## RMF analysis

Behavioral segmentation by 3 important features:

Recency  number of days since the last purchase

Frequency  number of transactions made over a given period

Monetary  amount spent over a given period of time

```python
from datetime import datetime
PRESENT = datetime(2018,9,3)
rfm= df.groupby('customer_unique_id').agg({'order_purchase_timestamp': lambda date: (PRESENT - date.max()).days,
                                        'order_id': lambda num: len(num),
                                        'payment_value': lambda price: price.sum()})
rfm.columns=['recency','frequency','monetary']
rfm['recency'] = rfm['recency'].astype(int)
rfm['frequency'] = rfm['frequency'].astype(int)
rfm['monetary'] = rfm['monetary'].astype(float)
```

```python
rfm.head()
```

```python
# Plot RFM distributions
plt.figure(figsize=(12,10))

# Plot distribution of R
plt.subplot(3, 1, 1)
sns.distplot(rfm['recency'],color='black')

# Plot distribution of F
plt.subplot(3, 1, 2)
sns.distplot(rfm['frequency'],color='black')

# Plot distribution of M
plt.subplot(3, 1, 3)
sns.distplot(rfm['monetary'],color='black')

plt.show()
```

- Above three plots are of recency, frequency, and monetary value. We see that customers stayed over longer with Olist which is positive thing but frequency of order is less.
- As per second and third plot, we find that order freqeuncy is less than 5, and amount spent is less 1500, respectively, over given period of time.

```python
# Create labels for Recency, Frequency and monetary
def partition(x):
    if x < 10:
        return 1
    if 10<=x<=35:
        return 2
    if 35<x<=50:
        return 3
    if 50<x<=75:
        return 4      

rfm['f_quartile']=rfm['frequency'].map(lambda cw : partition(cw) ) 
r_labels = range(4, 0, -1)
m_labels= range(1,5)

rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, r_labels)
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, m_labels)
```

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm.head()
```

```python
rfm_count_unique = rfm.groupby('RFM_Score')['RFM_Score'].nunique()
print(rfm_count_unique.sum())
```

```python
rfm['RFM_Score_s'] = rfm[['r_quartile','f_quartile','m_quartile']].sum(axis=1)
print(rfm['RFM_Score_s'].head())
```

```python
# Define rfm_level function
def rfm_level(df):
    if df['RFM_Score_s'] >= 9:
        return 'Can\'t Loose Them'
    elif ((df['RFM_Score_s'] >= 8) and (df['RFM_Score_s'] < 9)):
        return 'Champions'
    elif ((df['RFM_Score_s'] >= 7) and (df['RFM_Score_s'] < 8)):
        return 'Loyal'
    elif ((df['RFM_Score_s'] >= 6) and (df['RFM_Score_s'] < 7)):
        return 'Potential'
    elif ((df['RFM_Score_s'] >= 5) and (df['RFM_Score_s'] < 6)):
        return 'Promising'
    elif ((df['RFM_Score_s'] >= 4) and (df['RFM_Score_s'] < 5)):
        return 'Needs Attention'
    else:
        return 'Require Activation'

rfm['RFM_Level'] = rfm.apply(rfm_level, axis=1)

rfm.head()
```

```python
rfm_level_agg = rfm.groupby('RFM_Level').agg({
    'recency': 'mean',
    'frequency': 'mean',
    'monetary': ['mean', 'count']
}).round(1)
# Print the aggregated dataset
print(rfm_level_agg)
```
